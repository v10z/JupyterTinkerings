```python
import numpy
import math
import plotly.express as px
import numba
import cppyy
from pyevtk.hl import gridToVTK
```

```python
N = 500;
H_new = numpy.zeros((N,N,1));
p = 5;

H_new[int(N/2),int(N/2)] = 10**10;
t = 0;
tmax = 600;
p=0;
```

```python
def update_H(H_old):     
    points_above_threshold = H_old >= 4;

    H_old[points_above_threshold] -= 4;

    H_old[1:,:][points_above_threshold[:-1,:]] += 1;
    H_old[:-1,:][points_above_threshold[1:,:]] += 1;
    H_old[:,1:][points_above_threshold[:,:-1]] += 1;
    H_old[:,:-1][points_above_threshold[:,1:]] += 1;

    return H_old;
```

```python
def write_vtk(H,t):
    x = numpy.linspace(0, (N-1), N);
    y = numpy.linspace(0, (N-1), N);
    z = numpy.linspace(0, 1, 1);
    gridToVTK("./results/result_"+str(t), x, y, z, pointData = {"H":H});
```

```python
while (t<tmax):
    H_new=update_H(H_new);
    t+=1;
    if (t % 1 == 0):
        write_vtk(H_new,t)
        print(t)
```
